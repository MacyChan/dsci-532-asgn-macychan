```python
import pandas as pd
```

```python
dc_marvel = pd.read_csv('data/marvel_dc_characters.csv')
power_matrix = pd.read_csv('data/superheroes_power_matrix.csv')
```

```python
all_charaters = pd.merge(
    dc_marvel,
    power_matrix,
    how="inner",
    on="Name")


all_charaters.to_csv( 'characters_processed.csv', index=False)
```
