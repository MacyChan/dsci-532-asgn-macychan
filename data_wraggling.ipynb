```python
import pandas as pd
```

```python
dc_marvel = pd.read_csv('data/marvel_dc_characters.csv')
power_matrix = pd.read_csv('data/superheroes_power_matrix.csv')
char_stats = pd.read_csv('data/characters_stats.csv')
char_stats = pd.concat([char_stats.iloc[:,0:1], char_stats.iloc[:,2:]], axis=1)
```

```python
all_charaters = pd.merge(
    dc_marvel,
    char_stats,
    how="left",
    on="Name")
```
